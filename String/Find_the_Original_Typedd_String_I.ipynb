### **3333.Find the Original Typed String I**

Problem Link: https://leetcode.com/problems/find-the-original-typed-string-i/description/

#### **Approach**

**Step 1**: Group the consecutive characters

> Iterate through the string and group the same characters that appear together
> For example, "abbccc" â†’ group lengths: [1, 2, 3]

**Step 2**: Count how many possibilities we have

> Initialize total = 1 (the original word itself is one possibility)

> For each group of length k > 1, we can create:

> k - 1 new strings by shortening that group (to length 1 to k - 1)

> Since only one group can be shortened (reverse long press), we sum up all k - 1 from each group

**Step 3**: Add those possibilities to the total

> So final result = 1 + sum(length - 1 for each group with length > 1)

**CODE**

```python
class Solution:
    def possibleStringCount(self, word: str) -> int:
        if not word:
            return 0

        # Store lengths of all groups of repeated characters
        group_lengths = []

        i = 0
        while i < len(word):
            j = i
            while j < len(word) and word[j] == word[i]:
                j += 1

            group_lengths.append(j - i)  # Length of this group
            i = j

        total = 1  # Count the original string as one possibility


        # For each group, add (length - 1) shortened possibilities
        for length in group_lengths:
            total += (length - 1)

        return total
```

### **Time and  Space complexity**

> * Time complexity: Single pass through the string + loop over groups: **O(n)**

> * Space complexity: List to store group lengths, up to n groups: **O(n)**
