# 2215.Find the Difference of Two Arrays

Problem link: https://leetcode.com/problems/find-the-difference-of-two-arrays/description/?envType=study-plan-v2&envId=leetcode-75 


1. **Set-based solution**

```python
def findDifference(nums1, nums2):
    set1, set2 = set(nums1), set(nums2)
    return [list(set1 - set2), list(set2 - set1)]
```

##### Time Complexity

1. `set(nums1)` → **O(n)**
2. `set(nums2)` → **O(m)**
3. `set1 - set2` → **O(n)** (checking each element in `set1`)
4. `set2 - set1` → **O(m)**
5. `list(...)` conversion → **O(n + m)**

**Total: O(n + m)**

##### Space Complexity

* `set1` → O(n)
* `set2` → O(m)
* Result lists → O(n + m) in worst case

 **Total: O(n + m)**


2. **Manual traversal solution**

```python
def findDifference(nums1, nums2):
    res1, res2 = [], []
    
    for x in nums1:                 # O(n)
        if x not in nums2:          # O(m) check each time
            if x not in res1:       # O(n) worst case
                res1.append(x)
    
    for y in nums2:                 # O(m)
        if y not in nums1:          # O(n) check each time
            if y not in res2:       # O(m) worst case
                res2.append(y)
    
    return [res1, res2]
```

##### Time Complexity

* First loop: O(n \* (m + n))
* Second loop: O(m \* (n + m))

 **Total: O(n·m + n² + m²)** → much worse than the set-based approach.

##### Space Complexity

* `res1` up to O(n)
* `res2` up to O(m)

 **Total: O(n + m)**


##### **Comparison**

* **Set-based** → ✅ Efficient: O(n + m) time, O(n + m) space
* **Manual traversal** → ❌ Inefficient: O(n·m + n² + m²) time, O(n + m) space
